```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataset = pd.read_csv('position_salary.csv')
```

```python
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
```

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X, Y)
```

```python
polynomial_regressor = PolynomialFeatures(degree=4)
x_polynomial = polynomial_regressor.fit_transform(X)
x_polynomial
```

```python
linear_poly_regressor = LinearRegression()
linear_poly_regressor.fit(x_polynomial, Y)
```

```python
plt.scatter(X, Y, color='red')
plt.plot(X, linear_regressor.predict(X), color='blue')
plt.title('Linear Regressor')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X, Y, color='red')
plt.plot(X, linear_poly_regressor.predict(x_polynomial), color='blue')
plt.title('Polynomial Regressor')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
x_grid = np.arange(min(X), max(X), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
y_preds = linear_poly_regressor.predict(polynomial_regressor.fit_transform(x_grid))

plt.scatter(X, Y, color='red')
plt.plot(x_grid, y_preds, color='blue')
plt.title('Polynomial Regressor (smoothed)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
